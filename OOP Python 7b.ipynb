# Barang Service

```python
import mysql.connector
import pandas as pd

class Barang_Service:

    def __init__(self, barang):  
        self.id_barang = barang.id_barang
        self.nama_barang = barang.nama_barang
        self.kuantitas = barang.kuantitas
        self.harga_beli = barang.harga_beli
        self.harga_jual = barang.harga_jual        
        
        conn = None

    @staticmethod
    def get_koneksi_db():
        conn = mysql.connector.connect(host="localhost" ,
                                            user="root",
                                            password="",
                                            database="warung")
        if conn.is_connected():
            print("Berhasil terhubung ke database")
        
        return conn

    def insert_data(self):
        koneksi_db = Barang_Service.get_koneksi_db()
        cursor = koneksi_db.cursor()
        
        sql = "INSERT INTO barang (nama_barang, kuantitas, harga_beli, harga_jual) VALUES (%s, %s, %s, %s)"
        val = (self.nama_barang, self.kuantitas, self.harga_beli, self.harga_jual)
        
        cursor.execute(sql, val)

        koneksi_db.commit()

        print("{} data ditambahkan".format(cursor.rowcount))
        Barang_Service.tampilkan_data()
        
    def update_data(self):
        koneksi_db = Barang_Service.get_koneksi_db()
        cursor = koneksi_db.cursor()
        
        sql = "UPDATE barang set nama_barang=%s, kuantitas=%s, harga_beli=%s, harga_jual=%s WHERE id_barang=%s"
        val = (self.nama_barang, self.kuantitas, self.harga_beli, self.harga_jual, self.id_barang)

        cursor.execute(sql, val)

        koneksi_db.commit()

        print("{} data diupdate".format(cursor.rowcount))
        Barang_Service.tampilkan_data()

    
    def delete_data(self, id_barang):
        koneksi_db = self.get_koneksi_db()
        cursor = koneksi_db.cursor()
        
        sql = "DELETE FROM barang WHERE id_barang=%s"
        val = (id_barang, )
        cursor.execute(sql, val)

        koneksi_db.commit()

        print("{} data dihapus".format(cursor.rowcount))
        Barang_Service.tampilkan_data()
    
    @staticmethod
    def tampilkan_data():
        koneksi_db = Barang_Service.get_koneksi_db()
        cursor = koneksi_db.cursor()

        sql = "SELECT * FROM barang"
        cursor.execute(sql)

        results = cursor.fetchall()

        for data in results:
            print(data)
            
    @staticmethod
    def tampilkan_dataframe():
        koneksi_db = Barang_Service.get_koneksi_db()
        cursor = koneksi_db.cursor()

        sql = "SELECT * FROM barang"
        
        df = pd.read_sql(sql, con=koneksi_db)
        
        display(df)


Barang_Service.tampilkan_dataframe()
```
